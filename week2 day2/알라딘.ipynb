```python
# 라이브러리 임포트
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import codecs
```

```python
# 크롬드라이버 파일 세팅
driver = webdriver.Chrome("../chromedriver.exe")

# 알라딘 서점 베스트셀러 페이지로 이동
driver.get('https://www.aladin.co.kr/shop/common/wbest.aspx?BranchType=1&start=we')

# 소스코드 가져오기
source = driver.page_source
```

```python
s1 = BeautifulSoup(source, "html.parser")
```

```python
s_book = s1.find_all("div", class_="ss_book_box")

len(s_book)

for i in s_book :
    result = s_book[0].find_all('li')

len(result)
```

```python
for i in range(1,4) :
    st = result[i].text
    st = st.replace('\n', '')
    st = st.replace('\t', '')
    print(st)
```

```python
# 위 코드를 50번 반복해서 1위 ~ 50위까지 출력하기.
for item in s_book :
    print(item.find_all("li")[1].text)
    print(item.find_all("li")[2].text)
    print(item.find_all("li")[3].text)
    print('\n')
```

```python
path = 'D:/jjw/data/week2 day1/books.txt'
f = codecs.open(path, encoding='utf-8', mode='w')

rank = 1
for i in range(len(s_book)) :
    f.write('%d위 도서 정보\r\n' %(rank))
    f.write(item.find_all("li")[1].text+'\r\n')
    f.write(item.find_all("li")[2].text+'\r\n')
    f.write(item.find_all("li")[3].text+'\r\n')
    f.write('\r\n')
    rank += 1
    
f.close()
```
