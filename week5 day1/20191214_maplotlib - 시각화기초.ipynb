### Matplotlib

- 파이썬 기반 데이터 분석 환경에서 가장 일반적으로 사용되는 시각화 라이브러리
- 하위 라이브러리인 'pyplot'을 통해 주요 기능을 사용 
- plotly, seaborn 등과 같은 라이브러리가 최근 경쟁력을 갖추고 있음
- pandas 객체의 기본적인 시각화 도구
- plot() 메서드를 통해 데이터를 시각화
- 종류
    - 라인 플롯(line plot)
    - 바 차트(bar chart)
    - 히스토그램(histogram)
    - 박스 플롯(box plot)
    - 스캐터 플롯(scatter plot)

### plt.plot(y)
- 기본 : 라인 플롯(line plot)
- 데이터가 시간, 순서 등에 따라 변화하는 흐름 파악
- x가 주어지지 않을 경우 자동 생성 :x = range(len(y))

```python
# 데이터 생성 : 5 ~ 6 개의 아이템을 가지는 리스트
# plot(data) : 기본적으로 y축의 값으로 매칭시킴
# x -> len(data) 만큼의 인덱스번호를 자동으로 부여함
# 기본 plot = line plot
plt.plot([1,3,4,2,6,10])
plt.show()
```

```python
# x tick(x축 값) 설정
# plot.plot(x, 6) : x와 y의 길이가 같아야 함
plt.plot([2,4,6,8,10,12], [1,3,4,2,6,10])
plt.show()
```

### 스타일 지정

- plt.plot([x], y, [fmt], **kwargs)
    - fmt : '[color][marker][line]'
        - 색상, 마커, 선 등에 대한 포맷 옵션 설정

```python
# 스타일 문자열 추가 : [fmt] => color + marker + line 순서대로 문자열 작성
# 생략된 요소는 기본값으로 적용
# 'o' : 라인 없이 큰 점(원모양) 으로 데이터 표시
# o x ^
plt.plot([1,2,3,4], [2,6,10,7], 'o')
plt.show()
```

```python
# o- : 마커는 동그라미, 라인은 실선
# fmt [색상][마커][선종류]
plt.plot([1,2,3,4], [2,6,10,7], 'o-')
plt.show()
```

```python
# o-- : 마커는 동그라미, 선은 점선
plt.plot([1,2,3,4], [2,6,10,7],'o--')
plt.show()
```

```python
# ro : r(빨간색), o(마커), :(동그란 점선)
plt.plot([1,2,3,4], [2,6,10,7],'ro:')
plt.show()
```

##### 색상 및 스타일 참조 문서
마커 : https://matplotlib.org/3.1.1/api/markers_api.html <br>
라인 : https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html <br>
색상 : https://matplotlib.org/3.1.0/gallery/color/named_colors.html <br>

```python
# 기타 스타일 파라미터 종류
# color : magenta(선 색상)
# linestyle : --(선의 모양)
# marker : ^(마커의 모양)
# markerfacecolor : orange(마커 자체의 색상)
# makersize : 20(마커의 크기)
# markeredgecolor : r(마커 테두리의 색상)
plt.plot([1,2,3,4], [2,6,10,7], color="magenta", linestyle='-.',
         marker='^', markeredgecolor='r', markerfacecolor='orange', markersize='20')
plt.show()
```

### 축 범위 지정

#### matplotlib.pyplot.axis 

- xmin, xmax, ymin, ymax = plt.axis([xmin, xmax, ymin, ymax])
    - xmin : x축 최소값
    - xmax : x축 최대값
    - ymin : y축 최소값
    - ymax : y축 최대값

#### matplotplib.pyplot.xlim / ylim

- plt.xlim(min, max)
- plt.ylim(min, max)

```python
# x축 최소값 = 0, x축 최대값 = 7
# y축 최소값 = 2, y축 최대값 = 20
plt.plot([1,2,3,4], [10,3,8,13])
plt.axis([0,7,2,20])
plt.show()
```

```python
# 선에 비해 지나치게 작은 플롯을 구성하면 일부 플롯이 잘려나간다
plt.plot([1,2,3,4], [10,3,8,13])
plt.axis([2,4,4,10])
plt.show()
```

```python
# x축, y축, 범위 설정(최소값, 최대값)
# x data = 10, 20, 30, 40
# y data = 1, 4, 9, 16
# 색상 : magenta
# 선굵기 : 5
# 선모양 : :
# 마커 : o
# 마커크기 : 10
# x축 범위 : 0, 50
# y축 범위 : -10, 30
plt.plot([10,20,30,40], [1,4,9,16], c='red', lw=3, ls='--', marker='o', ms=5)
plt.xlim(0, 50)
plt.ylim(-10,30)
plt.show()
```

```python
# grid 표시하기
# grid는 격자무늬로 표시할때는 plt.grid(True)를 추가로 실행하면 됩니다.
plt.plot([10,20,30,40], [1,4,9,16], c='red', lw=3, ls='--', marker='o', ms=5)
plt.xlim(0, 50)
plt.ylim(-10,30)
plt.title('그리드 적용')
plt.grid(True)
plt.show()
```

```python
import numpy as np
```

```python
# 0이상 10미만의 범위에서 0.5씩 증가하는 실수 생성
data = np.arange(0, 10, 0.5)
data # 20개
```

```python
# 데이터셋 1: x, y 데이터가 동일(y = x), 파란색 --모양 라인
plt.plot(data, data, 'b--')
```

```python
# 데이터셋 2: y = x ** 2, 빨간색 네모 마커
plt.plot(data, data**2, 'r--')
```

```python
# 데이터셋 3 : y = x**3 녹색(g) 삼각형(^)
plt.plot(data, data**3, 'g^')
plt.show()
```

```python
plt.plot(data, data, 'b--')
plt.plot(data, data**2, 'rs')
plt.plot(data, data**3, 'g^')
plt.show()
```

```python
# data를 활용하여 총 3개의 data를 생성하여 하나의 plot으로 시각화
# plt.plot(xdata1, ydata1, fmt1, xdata2, ydata2, fmt2, xdata3, ydata3, fmt3...)
plt.plot(data, data, 'b-', data, data**2, 'r-', data, data**3, 'g-')
plt.show()
```

#### matplotlib.pyplot.scatter 

- 대표값을 중심으로 데이터가 흩어져 있는 정도를 시각화
- 데이터의 분산 정도, 데이터 간의 상관관계를 파악할 수 있음
- plt.scatter(x, y, s=None, c=None, data=None)
    - s : 마커 사이즈
    - c : 마커 색상
    - data : dictionary 자료형을 사용하여 데이터를 전달하면 keyword string을 통해 데이터 사용 가능

```python
# 정규 분포 확률 밀도에서 100개의 표본 추출
# random.normal()
np.random.seed(123)
x = np.random.normal(0,1,100)
y = np.random.normal(0,1,100)
plt.scatter(x,y)
plt.show()
```
