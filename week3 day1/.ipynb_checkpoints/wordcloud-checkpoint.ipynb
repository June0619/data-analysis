# konlpy를 활용한 형태소 분석

- 형태소란 단어를 더 이상 나눌 수 없는 단위로 나눠놓은 단어입니다.

- konlpy는 한국어 단어를 더 나눌 수 없는 단위로 쪼갤 수 있는 라이브러리입니다.

- 우리가 통상적으로 사용하는 한국어 문장을 형태소 단위로 나누는 작업이 선행되어야 워드클라우드를 사용할 수 있습니다.

- konlpy document 주소(https://konlpy-ko.readthedocs.io/ko/v0.4.3/)

```python
# konlpy의 꼬꼬마 라이브러리 사용
from konlpy.tag import Kkma
```

```python
# 꼬꼬마 라이브러리 변수(객체) 생성
kkma = Kkma()
```

```python
# 문장단위 입력
kkma.sentences('너무 피곤해 집에 가고싶다.')
```

```python
# 명사 쪼개기
kkma.nouns('너무 피곤해서 집에 가고싶은데 수업이 네 시간이나 남았다.')
```

```python
# 전체 문장을 형태소 단위로 쪼개기
kkma.pos('너무 피곤해서 집에 가고싶은데 수업이 네 시간이나 남았다.')
```

```python
# 한나눔 라이브러리를 사용하기
from konlpy.tag import Hannanum
```

```python
# 한나눔 변수(객체) 생성
hannanum = Hannanum()
```

```python
# 한나눔에서 명사 쪼개기
hannanum.nouns('너무 피곤해서 집에 가고싶지만 수업이 네 시간이나 남았다')
```

```python
# 품사표시 없이 문장 쪼개기
hannanum.morphs('너무 피곤해서 집에 가고싶지만 수업이 네 시간이나 남았다.')
```

```python
# 품사표시와 함께 문장 쪼개기
hannanum.pos('포장마차 그때 그 거리에서 그때 그곳에 서서 날 사랑한다 말했잖아, 영원할거라고 말했잖아')
str1 = '포장마차 그때 그 거리에서 그때 그 곳에 서서 날 사랑한다 말했잖아, 영원할거라고 말했잖아'
```

```python
# Twitter 분석기 사용
from konlpy.tag import Twitter
```

```python
# Twitter 변수(객체) 생성
t = Twitter()
```

```python
# 명사분석
print(str1)
t.nouns(str1)
```

```python
# 형태소 분류
t.morphs(str1)
```

```python
# 형태소, 품사 분류
t.pos(str1)
```

# 워드 클라우드

- 워드 클라우드는 문장이나 글을 입력했을때 형태소 단위로 분석한 다음, 빈출단어 순으로 글꼴 크기를 다르게 해서 이미지화합니다.

- 시각적으로 빈출단어와 그렇지 않은 단어를 확인할 수 있도록 만들어주기 때문에 분석결과를 직관적으로 파악할 수 있습니다.

- 워드클라우드 공식 페이지(https://amueller.github.io/word_cloud/ )를 참조해 워드클라우드 사용법을 파악해보겠습니다.

```python
from wordcloud import WordCloud, STOPWORDS

import matplotlib.pylab as plt

import numpy as np
from PIL import Image
```

### 스타워즈 시각화

- 스타워즈 대본 및 스톰 트루퍼 사진을 활용한 워드클라우드화 및 그냥 시각화 두가지 진행

- 스톰 트루퍼 사진(흑백화 사진)을 사용하면 스톰 트루퍼 사진의 검은색 영역에만 워드 클라우드가 형성됨.

<img src="stormtrooper.png" width="350" height="350">

```python
# 스타워즈 txt 내부의 탁스트 자료를 파이썬 문자열 자료형으로 변환
starwars = open('starwars.txt', encoding='utf-8').read()
```

```python
# 배경자료를 줄 지 말지 선택( 입력 안하면 배경 지정 없음 )
background = np.array(Image.open('stormtrooper.png'))
```

```python
# 워드클라우드를 만들 때 제외할 단어 입력하기
stopwords = set(STOPWORDS)
stopwords.add('int')
stopwords.add('ext')
```

```python
# 시각화 툴에 자료 집어넣기
# background_color : 배경색
# max_words : 최대 몇 글자로 클라우드 구성
# mask : 배경에 사용할 그림파일 정보
# stopwords : 워드클라우드 생성시 배제할 단어
# margin : 그림 모서리 부분과 그림 시작점의 간격
# random_state : 그림 구성(숫자를 다르게 주면 단어 배치만 조금씩 바뀜)
# .generate(영문자료) : 입력된 영문자료의 형태소 분석 + 워드클라우드화를 동시에 진행

wc = WordCloud(background_color="black", max_words=1000, mask=background,
              stopwords=stopwords, margin=10, random_state=100).generate(starwars)
#글씨 색깔 지정
default_colors = wc.to_array()
```

### imshow의 interpolation

- 경계를 어떻게 표시할지 지정할때 interpolation을 사용함

https://matplotlib.org/3.1.1/gallery/images_contours_and_fields/interpolation_methods.html

- 위 사이트에 옵션별 경계면 처리를 좀 더 상세히 확인 가능

```python
# 시각화한 자료 화면에 표출하기
# 12 * 12 크기로 출력
plt.figure(figsize=(12, 12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# 뒷 배경이 검은색이며, 배경그림이 ko.png인 워드클라우드를 직접 만들기
text_file = open('./news_list/news.txt', encoding='utf-8').read()

background = np.array(Image.open('kor_map.png'))

stopwords = set(STOPWORDS)
stopwords.add('co')
stopwords.add('kr')

wc = WordCloud(background_color="black", max_words=1000, mask=background,
              stopwords=stopwords, margin=10, random_state=5).generate(starwars)

default_colors = wc.to_array()
```

## 한국어를 이용한 워드 클라우드

```python
import nltk

# kobill은 한국어로 된 법조문 파일 10개를 담고있는 라이브러리
from konlpy.corpus import kobill
```

```python
# 법조문 파일 10개 로드
corp=kobill.fileids()
corp
```

```python
doc_ko = kobill.open('1809899.txt').read()
doc_ko
```

```python
# 트위터 라이브러리를 이용한 명사만 가져오기
token = t.nouns(doc_ko)
token
```

```python
# 한국어 워드클라우딩은 단어 카운팅이 선행되어야 함.
count_voca = nltk.Text(token, name="단어카운팅")
print(len(count_voca.tokens))
print(len(set(count_voca.tokens)))
count_voca.vocab()
```

```python
# 카운팅 된 단어 중 최빈출 단어 150개만 남겨서 워드클라우딩
voca = count_voca.vocab().most_common(150)
# 한국어 폰트가 워드클라우드에 내장되어있지 않으므로 글꼴 지정
font_path = 'c:/Windows/Fonts/malgun.ttf'
# 한국어 워드클라우딩은 영문과 약간 다르게 진행됨
# font_path : "글꼴 경로"
# relative_scaling : 빈출단어의 단어 크기를 어느정도로 잡을것인지
# .generate_from_frequencies(dict(빈도산출)) : 워드클라우드 생성

wordcloud = WordCloud(font_path=font_path,
                     relative_scaling=0.2,
                     background_color='black').generate_from_frequencies(dict(voca))
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
